# example of data analysis

This is an example of data analysis using numpy array and matplotlib histogram. The data were stored in a csv (comma-separated values)file and read into a <b>pandas</b> data frame in the python code. The structure of the data is a row-column format, similar to Exel data sheet, and each column is identified by a key word, for example, "Volume", "Weight" and "CO2", as seen below. The kew words are used to extract data from the pandas data frame to numpy array.
<br>
The csv file was from<br>
https://www.w3schools.com/python/python_ml_multiple_regression.asp
<br>
To try this program, download the csv file "data_w3_car_co2.csv" from this github page.<br>
https://github.com/skunori/phys2305s23/blob/main/jn_examples/data_w3_car_co2.csv 
<br>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas

df = pandas.read_csv("data_w3_car_co2.csv")
print(df)
```

```python
volumes=df["Volume"].to_numpy()   # extract Volume data
print(volumes)
```

```python
weights=df["Weight"].to_numpy()   # extract Weight data
print(weights)
```

```python
co2s=df["CO2"].to_numpy()        # extract CO2 data
print(co2s)
```

```python
# make histograms
plt.hist(volumes)
plt.show()
```

```python
plt.hist(weights)
plt.show()
```

```python
plt.hist(co2s)
plt.show()
```

```python
#  look at correlation
plt.scatter(volumes,weights)
plt.xlabel("volumes")
plt.ylabel("weights")
plt.show()
```

```python
plt.scatter(volumes,co2s)
plt.xlabel("volumes")
plt.ylabel("co2s")
plt.show()
```

```python
plt.scatter(weights,co2s)
plt.xlabel("weights")
plt.ylabel("co2s")
plt.show()
```

```python
# How to deal with character string data?
cars=df["Car"]
plt.scatter(co2s,cars)
plt.xlabel("CO2")
plt.show()
```

#  preview of ML

The following is from <b>Machine Learning - Multiple Regression</b> at<br>
https://www.w3schools.com/python/python_ml_multiple_regression.asp
<br><br>
We import the sklearn module for this ML, which builds a model to predict the CO2 production for a set of paramters, [Weight, Volume] based on training data. <br><br>
From the sklearn module we will use the LinearRegression() method to create a linear regression object. <br><br>
This object has a method called fit() that takes the independent and dependent values as parameters and fills the regression object with data that describes the relationship. <br><br>
Then, we have a regression object that are ready to predict CO2 values based on a car's weight and volume:<br>

```python
# preview of ML
from sklearn import linear_model

X = df[['Weight', 'Volume']]
y = df['CO2']

# print(X)
# print(y)

regr = linear_model.LinearRegression()
regr.fit(X.values, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[2300, 1300]])

print(predictedCO2,"predicted CO2")
```

```python
#  projection of points:  

xx=df["Weight"].to_numpy()
yy=df["Volume"].to_numpy()
zz=df['CO2'].to_numpy()
print(xx)
print(yy)
print(zz)
xval=np.array([2300.0])
yval=np.array([1300.0])
zval=np.array(predictedCO2)
print("xval",xval)
print("yval",yval)
print("zval",zval)

plt.scatter(xx,zz)
plt.scatter(xval,zval)
plt.xlabel("Weight")
plt.ylabel("CO2")
plt.xlim(500.0,2500.0)
plt.show()

plt.scatter(yy,zz)
plt.scatter(yval,zval)
plt.xlabel("Volume")
plt.ylabel("CO2")
plt.xlim(500.0,2500.0)
plt.show()
```

```python
#  The following is a magic line for interctive 3D rotation of the plot
%matplotlib notebook

fig = plt.figure()
ax = fig.add_subplot(projection='3d')
#ax.scatter(xx,yy,zz)
ax.scatter(xx,yy,zz)
ax.scatter(xval,yval,zval,marker="X")
ax.set_xlabel("Weight")
ax.set_ylabel("Volume")
ax.set_zlabel("CO2")
ax.set_xlim(500.0,2500.0)
ax.set_ylim(500.0,2500.0)
ax.set_zlim(80.0,130.0)
plt.show()
```
