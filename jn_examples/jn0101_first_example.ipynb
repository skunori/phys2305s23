```python
import numpy as np
import matplotlib.pyplot as plt

x=np.arange(0.0, 10.0, 0.01)   # values from 0.0 to 10.0 with 0.01 step in numpy array
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y)   # plot sin curve
plt.plot(x,z)   # plot cos curve
plt.show()
```
