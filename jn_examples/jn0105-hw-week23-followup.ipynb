```python
#   example-1
import numpy as np
import matplotlib.pyplot as plt

def myfunction(x):
    print("function myfunction is called.")
    y=np.sin(x)/x-0.05 
    return y

# main program
print("before x=")
x=np.arange(-10.0,10.0,0.1) 
print("before y=")
y=myfunction(x)
print("after y=")

xzero=np.array([-10.0,10.0]) 
yzero=np.array([0.0,0.0])
plt.plot(x,y,'-b')
plt.plot(xzero,yzero,'-r') 
plt.show()
```

```python
#   example- week3
import numpy as np
import matplotlib.pyplot as plt

def myfunction(x):
   y=2.0*x*x+3.0*x+4.0
   return y


def rectangular(dx,xmin,xmax):
    print("function rectangular is called")
    print("dx=",dx)
    ysum=0.0
    xx=np.arange(xmin,xmax,dx)
    for x in xx:
        a=x
        b=x+dx
        c=(a+b)/2.0
        y=dx*myfunction(c)
        ysum=ysum+y
    return ysum

def trapezoidal(dx,xmin,xmax):
    print("function trapezoidal is called")
    ysum=0.0
    return ysum

def simpsons(dx,xmin,xmax):
    print("function trapezoidal is called")
    ysum=0.0
    return ysum
   

dx=0.01
xmin=-5.0
xmax=5.0

# three methods to integrate...
print("before result1=")
result1=rectangular(dx,xmin,xmax)
print("before result2=")
result2=trapezoidal(dx,xmin,xmax)
print("before result3=")
result3=simpsons(dx,xmin,xmax)

print('rectangular ',result1)
print('trapezoidal ',result2)
print('simpsons    ',result3)

#  display the sahpe of function.
x=np.arange(xmin,xmax,dx)
y=myfunction(x)
fig=plt.figure(1)
plt.plot(x,y,'g')
fig.text(0.2,0.8,'F(x)=2.0*x*x+3.0*x+4.0',ha='left',va='center')
r1='rectangular: '+str(result1)
fig.text(0.3,0.70,r1,ha='left',va='center')
fig.savefig("myintegrationA.png")
plt.show()
```
